```python
from acgan import ACGAN
import cv2
```

```python
gan = ACGAN()
```

```python
d_losses, g_losses = gan.train(epochs=10000, batch_size=256, sample_interval=200)
```

```python
gan.sample_images(1000)
import numpy as np
g_losses, d_losses = np.array(g_losses), np.array(d_losses)
```

```python
import matplotlib.pyplot as plt
plt.plot(g_losses[:,0], label='generator loss')
plt.plot(d_losses[:,0], label='discriminator loss')
plt.legend()
plt.show()
```
