````python
import numpy as np

X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)

plt.plot(X, C)
plt.plot(X, S)
plt.show()
````

````python
plt.plot(X, C, color="blue", linewidth=1)
plt.plot(X, S, color="red",  linewidth=2.5, linestyle="--")
plt.show()
````

````python
plt.plot(X, C, color="blue", linewidth=1, label='cos')
plt.plot(X, S, color="red",  linewidth=2.5, linestyle="--", label='sin')

plt.xlim(X.min()*1.1, X.max()*1.1)
plt.ylim(C.min()*1.1, C.max()*1.1)

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, +1])

plt.legend()

plt.show()
````

# Exercise on Numpy and Matplotlib

The data in [populations.txt](http://www.scipy-lectures.org/_downloads/populations.txt) describes the populations of hares and lynxes (and carrots) in northern Canada during 20 years:

Computes and print, based on the data in populations.txt...

* The mean and std of the populations of each species for the years in the period.
* Which year each species had the largest population.
* Which species has the largest population for each year. (Hint: argsort & fancy indexing of np.array([‘H’, ‘L’, ‘C’]))
* Which years any of the populations is above 50000. (Hint: comparisons and np.any)
* The top 2 years for each species when they had the lowest populations. (Hint: argsort, fancy indexing)
* Compare (plot) the change in hare population (see np.gradient) and the number of lynxes. Check correlation (see np.corrcoef).

... all without for-loops.

````python
!wget -r http://www.scipy-lectures.org/_downloads/populations.txt
````

````python

data = np.loadtxt('./populations.txt')
years   = data[:, 0]
hare    = data[:, 1]
lynx    = data[:, 2]
carrots = data[:, 3]
print("mean of hare is {}, std of hare is {}".format(hare.mean(), hare.std()))
print("mean of lynx is {}, std of lynx is {}".format(lynx.mean(), lynx.std()))
print("mean of carrots is {}, std of carrots is {}".format(carrots.mean(), carrots.std()))

print("max of hare was in {}".format(years[hare.argmax()]))
print("max of lynx was in {}".format(years[lynx.argmax()]))
print("max of carrots was in {}".format(years[carrots.argmax()]))

print(np.concatenate([
    years.reshape((-1, 1)), 
    np.array(['H', 'L', 'C'])[data[:,1:].argmax(axis=1)].reshape((-1, 1))
], axis=1))

print("for hare {}, for lynx {}, for carrots {}".format(years[hare>=50000], years[lynx>=50000], years[carrots>=50000]))

print("the top 2 year for hares {}".format(years[np.argsort(hare)[0:2]]))
print("the top 2 year for lynx {}".format(years[np.argsort(lynx)[0:2]]))
print("the top 2 year for carrots {}".format(years[np.argsort(carrots)[0:2]]))

plt.plot(years, lynx)

plt.plot(years, np.gradient(hare))
plt.show()

print("correlation is {}".format(np.corrcoef(np.gradient(hare), lynx)))
````
